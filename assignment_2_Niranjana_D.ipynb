TASK 3

** UNIVARIENT ANALYSIS

```python
## univarient analysis

sns.distplot(df.Date)
```

```python
sns.displot(df1.Price)  ### dis plot
```

```python
## piechart -- univrient

plt.pie(df1.Price.value_counts(),[0,0.1,0,0.1,0.1],labels=[0,1,2,3,4],shadow=True,
        colors=['pink','lightblue','orange','lightgreen','yellow'])
plt.title('PRICE')
plt.show
```

```python
#### barplot -- univarient

sns.barplot(x=df1.Price.value_counts().index,y=(df1.Price.value_counts()))
```

** BIVARIENT ANALYSIS

```python
###   Bi varient Analysis  --- scatter plot , line plot , joint plot
```

```python
sns.lineplot(x= df.Date ,y= df.Price)  ## lineplot -- bivarient
```

```python
sns.scatterplot(x= df.Date,y= df.Price)  ## scatterplot -- bivarient
```

```python
## bivarient
sns.jointplot(x='Date',y='Lattitude',data=df)
```
